```python
df.dropna(inplace=True)
```

```python
df["word_count"]=df['review_content'].apply(lambda x: len(str(x).split(" ")))
df[['word_count', 'review_content']].head(5)
```

```python
df["character_count"]=df["review_content"].str.len()
```

```python
df
```

```python
def avg_word(sentence):
    words=sentence.split()
    return(sum(len(word) for word in words)/len(words))
df["word_length"]=df["review_content"].apply(lambda x: avg_word(x))
df[["word_length", "review_content"]].head()
```

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')

df["stop_words"]=df["review_content"].apply(lambda x: len([x for x in x.split() if x in stop]))
df[["stop_words", "review_content"]].head()
```

```python
df["numerics"]=df["review_content"].apply(lambda x: len([x for x in x.split() if x.isdigit()]))
df[["numerics", "review_content"]].head()
```

```python
df["upper"]=df["review_content"].apply(lambda x: len([x for x in x.split() if x.isupper()]))
df[["upper", "review_content"]].head()
```

```python
# convert to lower
df['review_content'] = df['review_content'].apply(lambda x: (" ".join(x.lower() for x in x.split())))
df['review_content'].head(10)
```

```python
#remove non space and non word char
df["review_content"]=df["review_content"].str.replace('[^\w\s]', '')
df["review_content"]
```

```python
from nltk.corpus import stopwords
stop=stopwords.words('english')

df["review_content"]=df["review_content"].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
df["review_content"].head(10)
```

```python
# Frequently repeated words
freq=pd.Series(' '.join(df["review_content"]).split()).value_counts()[:20]
freq
```

```python
freq=list(freq.index)
freq
```

```python
#infrequent
freq=pd.Series(" ".join(df["review_content"]).split()).value_counts()[-10:]
freq
```

```python
#get words not in frequent
df["review_content"]=df["review_content"].apply(lambda x: " ".join(x for x in x.split() if x not in freq))
df["review_content"].head(10)
```
