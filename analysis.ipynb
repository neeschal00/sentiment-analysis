```python
df["word_count"]=df['review_content'].apply(lambda x: len(str(x).split(" ")))
df[['word_count', 'review_content']].head(5)
```

```python
df["character_count"]=df["review_content"].str.len()
```

```python
df
```

```python
def avg_word(sentence):
    words=sentence.split()
    return(sum(len(word) for word in words)/len(words))
df["word_length"]=df["review_content"].apply(lambda x: avg_word(x))
df[["word_length", "review_content"]].head()
```

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')

df["stop_words"]=df["review_content"].apply(lambda x: len([x for x in x.split() if x in stop]))
df[["stop_words", "review_content"]].head()
```
