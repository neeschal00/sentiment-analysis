```python
grade = {
    "A+":100,
    "A": 95,
    "A-": 90,
    "B+": 88,
    "B": 85,
    "B-": 80,
    "C+": 78,
    "C":75,
    "C-":70,
    "D+": 68,
    "D": 65,
    "D-": 60,
    "F": 40
}
    
```

```python
def convertP(value):
    divPattern = r"(\d+.)?\d+?\/\d+"
    floatPat = re.match(divPattern, value)
    if floatPat:
        dividend,divisor= value.split("/")
        percentValue = (float(dividend)/float(divisor))*100
        if(percentValue > 100):
            return None
        return int(percentValue)
    elif re.match(r"\w{3}-\d+",value):
        rating = value.split('-')[1]
        return int(rating)
    elif re.match(r"[A-F][\+\-]?",value):
        return int(grade[value])
    elif re.match(r"\d{3}",value):
        return int(value)/10
    elif re.match(r"\d{2}",value):
        return int(value)
    elif re.match(r"\d\-\w{3}",value):
        return None
    
    elif re.match(r"\d{1}",value):
        return (int(value)/10)*100
    else:
        return None

print(convertP("Feb-54"))
# new_df['review_score'] = new_df['review_score'].apply(convertP)
# new_df
```
